## TASK 1: Scrap Data from website and create Dataframe

### Start of Task 1

```python
import requests
from bs4 import BeautifulSoup as bs

import pandas as pd
# The following allows it to print the whole DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

import numpy as np
```



Sending a GET request to the URL.


```python
website_url = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&diff=945633050&oldid=942851379').text
```

```python
soup = bs(website_url, 'lxml')
```

To view the tags nested in the document.

```python
print(soup.prettify())
```

Find the table in the data of the page.

```python
my_table = soup.find("table", {"class": "wikitable sortable"})
```

**Visualise the table in HTML form.**

```python
my_table
```

```python
rows=0
n_col=0
col_names=[]
```

Using rows variable to find the number of rows in the table.

Looping through the table (HTML form) to find the column names for the pandas DataFrame.

```python
for row in my_table.find_all('tr'):
    tags = row.find_all('th') 
    for th in tags:
        col_names.append(th.get_text().strip('\n'))
    rows+=1
        
col_names
rows-=1
```

Displaying number of rows.

```python
rows
```

Declaring a pandas DataFrame with specified rows and columns with default value = NaN.

```python
df = pd.DataFrame(columns = col_names,index= range(0,rows))
```

**Extracting data from the table (HTML form) and transferring it to df (pandas DataFrame).**

```python
n_row=0
for row in my_table.find_all('tr'):
    n_col = 0
    col = row.find_all('td')
    for c in col:
        df.iat[n_row,n_col] = c.get_text().strip('\n')
        n_col += 1
    if len(col) > 0:
        n_row += 1

df
```

**Removing the rows for which Borough is 'Not assigned'.**

```python
df = df[df.Borough != 'Not assigned']
```

```python
df
```

If a cell has a Borough assigned but not the Neighbourhood is **Not Assigned**, it is assigned the **same value** as the **Borough**. 
The following code effectuates that:

```python
for i in range(210):
    if(df.iat[i, 2] == 'Not assigned'):
        # print(df.iat[i, 2])
        df.iat[i, 2] = df.iat[i, 1]
        # print(df.iat[i, 2])

df
```

Creating new DataFrame **df_new** for storing the final processed DataFrame. 

```python
df_new = pd.DataFrame(columns = col_names)
```

Below, I have clubbed all the Neighbourhood values for the same Postal Code. 

**Procedure:**
1. Generate a list of all unique values in the column **Postal Code**.
2. Loop through the DataFrame using these unique values.
3. Store the rows that have the same Postal Codes into another DataFrame: **df_d**.
4. Take all the column values of the Neighbourhood column of this new DataFrame **df_d** and store them in a list. Convert this list into a string. 
5. Now, generate a dictionary with all the values for a row of the DataFrame **df_new** in the following pattern. Example: 
     - {'Postcode': 'M6A',
     - 'Borough': 'North York',
     - 'Neighbourhood': 'Lawrence Heights, Lawrence Manor'}
6. Append this new row into the new DataFrame **df_new**. 

```python
for code in df.Postcode.unique():
    df_d = df[df.Postcode == code]
    list1 = df_d['Neighbourhood'].tolist()
    # print(list1)
    list1 = ', '.join(list1)
    new_row = {'Postcode': df_d.iat[0, 0], 'Borough': df_d.iat[0, 1], 'Neighbourhood': list1}
    # print(new_row)
    df_new = df_new.append(new_row, ignore_index = True)
```

Thus, the final processed DataFrame.

Defining a funcion that will return a DataFrame consisting of all the nearby Venues found for each neighbourhood.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### Ananlyse each neighbourhood

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighbourhood'] = toronto_venues['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

New DataFrame size is as follows

```python
toronto_onehot.shape
```

#### Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
toronto_grouped
```

Size of the new DataFrame **toronto_grouped**

```python
toronto_grouped.shape
```

#### Let's print each neighborhood along with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### Let's put that into a pandas dataframe

First, let's write a function to sort the venues in descending order.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Now let's create the new dataframe and display the top 10 venues for each neighborhood.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
venues_sorted = pd.DataFrame(columns=columns)
venues_sorted['Neighbourhood'] = toronto_grouped['Neighbourhood']

for ind in np.arange(toronto_grouped.shape[0]):
    venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

venues_sorted.head()
```

### Cluster Neighbourhoods

Run k-means to cluster the neighborhood into 5 clusters.

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighbourhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_ 
# to change use .astype()
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = df_toronto

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')

toronto_merged.head() # check the last columns!
```

In case, data not available for some row, we do the following:

```python
toronto_merged = toronto_merged.dropna()
```

Changing all Cluster labels from double to int type

```python
toronto_merged['Cluster Labels'] = toronto_merged['Cluster Labels'].astype('int64')
toronto_merged
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
